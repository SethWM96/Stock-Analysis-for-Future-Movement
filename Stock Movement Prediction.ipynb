```python
google = yahooFinance.Ticker('GOOG')
```

```python
df = google.history('5Y')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(7,5),dpi=150)
plt.title('Closing Price')
plt.plot(df['Close'])
```

```python
df['FiftyDMA'] = df['Close'].rolling(50).mean()
df['TwoHundredDMA'] = df['Close'].rolling(200).mean()
df.describe()
```

```python
df.drop(columns = ['Dividends','Stock Splits'], inplace=True)
```

```python
df.tail()
```

```python
df.head()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(7,5),dpi=150)
plt.title('Closing Price 50DMA vs 200DMA')
plt.plot(df['Close'],label = 'Close')
plt.plot(df['FiftyDMA'],label = 'FiftyDMA')
plt.plot(df['TwoHundredDMA'],label = 'TwoHundredDMA')
plt.legend()
```

## Inferences from the above graph
1. 50DMA and 200DMA are fairly good indicators of how the stock is moving.
2. If the closing price is above the 50DMA or 200DMA, it is usually bullish. 
3. If the closing price is below the 50DMA or 200DMA, it is usually on a down trend. 
4. Where 50DMA and 200DMA intersect, the market usually reverses its trend.
